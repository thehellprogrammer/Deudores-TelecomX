```python
import pandas as pd
```

```python
datos_TelecomX = pd.read_json('TelecomX_Data.json')
```

#ðŸ”§ TransformaciÃ³n

```python
customer = pd.json_normalize(datos_TelecomX['customer'])
```

```python
phone = pd.json_normalize(datos_TelecomX['phone'])
```

```python
internet = pd.json_normalize(datos_TelecomX['internet'])
```

```python
account = pd.json_normalize(datos_TelecomX['account'])
```

```python
Unido = pd.concat([columna_extraida ,customer, phone, internet, account], axis=1)

Deudores =Unido[Unido['Churn'] == 'Yes']
NoDeudores =Unido[Unido['Churn'] == 'No']
#Unido_female = Unido[Unido['gender'] == 'Female']
#Unido_Male = Unido[Unido['gender'] == 'Male']
Deudores.info()
NoDeudores.info()
```
